```python
# Step 1: Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Create a sample dataset
data = pd.DataFrame({
    'Age': [23, 25, 31, 22, 30, 29, 40, 36, 28, 27],
    'Salary': [50000, 52000, 58000, 49000, 56000, 54000, 60000, 62000, 51000, 53000]
})

# Print the dataset
print("Dataset:")
print(data)

# Step 3: Calculate descriptive statistics
mean_values = data.mean()
median_values = data.median()
std_values = data.std()

print("\nMean:")
print(mean_values)

print("\nMedian:")
print(median_values)

print("\nStandard Deviation:")
print(std_values)

# Step 4: Visualize data distributions using histograms
data.hist(bins=5, figsize=(10, 5), color='skyblue', edgecolor='black')
plt.suptitle('Data Distributions')
plt.show()
```
